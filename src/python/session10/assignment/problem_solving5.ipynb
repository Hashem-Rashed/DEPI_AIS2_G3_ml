```python
class Solution:
    def reverseWords(self, s: str) -> str:
        
        words = s.split()        # splits by any whitespace & removes extras
        return " ".join(words[::-1])
```

```python
def productExceptSelf(nums):
    n = len(nums)
    answer = [1] * n

    # Prefix products
    prefix = 1
    for i in range(n):
        answer[i] = prefix
        prefix *= nums[i]

    # Suffix products
    suffix = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= suffix
        suffix *= nums[i]

    return answer
```
