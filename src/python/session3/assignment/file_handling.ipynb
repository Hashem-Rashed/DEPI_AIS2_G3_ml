```python
import os
import random

def manage_folder(folder_path, num_files):
    """
    1. Open/create folder if it doesn't exist.

    2. Create `num_files` empty files inside it.

    3. Check and print number of files.

    4. Delete a random half of the files.

    5. Check and print remaining files.
    
    """
    
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    
    for i in range(1, num_files + 1):
        file_path = os.path.join(folder_path, f"file_{i}.txt")
        with open(file_path, "w") as f:
            f.write(f"This is file {i}\n")
    
    files = os.listdir(folder_path)
    print(f"Number of files before deletion: {len(files)}")
    
    num_to_delete = len(files) // 2
    files_to_delete = random.sample(files, num_to_delete)
    for file_name in files_to_delete:
        os.remove(os.path.join(folder_path, file_name))
    
    remaining_files = os.listdir(folder_path)
    print(f"Number of files after deletion: {len(remaining_files)}")
    print("Remaining files:", remaining_files)

folder = "test_folder"
manage_folder(folder, 10)
```
