# task 1

```python
class Fibonacci:
    """
    Class to generate Fibonacci sequence up to n elements.
    """
    def __init__(self, n):
      
        self.n = n

    def generate(self):
      
        def fib(k):
            if k == 0: return 0
            if k == 1: return 1
            return fib(k - 1) + fib(k - 2)
        return [fib(i) for i in range(self.n)]

# Example

fib = Fibonacci(10)
print(fib.generate())
```

```python
class SetOperations:
    """
    Class to perform operations on a set, like removing items.
    """
    def __init__(self, my_set):
        self.my_set = my_set

    def remove_items(self, items):
      
        self.my_set.difference_update(items)
        return self.my_set

# Example

s = SetOperations({1, 2, 3, 4, 5})
print(s.remove_items({1, 4}))
```

# task 3

```python
class TupleOperations:
    """
    Class to operate on a tuple, e.g., finding index of an item.
    """
    def __init__(self, tpl):
        self.tpl = tpl

    def index_of(self, item):
     
        if item in self.tpl:
            return self.tpl.index(item)
        else:
            return -1  

# Example

t = TupleOperations(('apple', 'banana', 'orange'))
print(t.index_of('orange'))
```

```python
class Sentence:
    """
    Class to handle sentence operations.
    """
    def __init__(self, text):
        self.text = text

    def to_words(self):
      
        return self.text.split()

# Example

s = Sentence("Welcome to Amit")
print(s.to_words())
```

```python
class Palindrome:
    """
    Class to check if a string is a palindrome.
    """
    def __init__(self, text):
        self.text = text

    def is_palindrome(self):
      
        return self.text == self.text[::-1]

# Example

p = Palindrome("racecar")
print("Palindrome" if p.is_palindrome() else "Not a Palindrome")
```

```python
class Factorial:
    """
    Class to calculate factorial of a number.
    """
    def __init__(self, n):
        self.n = n

    def calculate(self):
        """
        Calculates factorial using recursion.
        """
        if self.n == 0 or self.n == 1:
            return 1
        else:
            return self.n * Factorial(self.n - 1).calculate()

# Example

f = Factorial(5)
print(f.calculate())
```

# task 7

```python
class Squares:
    """
    Class to generate squares of numbers.
    """
    def __init__(self, n):
        self.n = n

    def generate(self):
      
        if self.n == 0:
            return []
        return Squares(self.n - 1).generate() + [self.n**2]

# Example

sq = Squares(10)
print(sq.generate())
```

# task 8

```python
class EvenSum:
    """
    Class to calculate the sum of even numbers in a list.
    """
    def __init__(self, numbers):
        self.numbers = numbers

    def calculate(self):
      
        return sum(num for num in self.numbers if num % 2 == 0)

# Example

e = EvenSum([1, 2, 3, 4, 5, 6])
print(e.calculate())
```

```python
class ReverseList:
    """
    Class to reverse a list.
    """
    def __init__(self, lst):
        self.lst = lst

    def reverse(self):
       
        if len(self.lst) == 0:
            return []
        return [self.lst[-1]] + ReverseList(self.lst[:-1]).reverse()

# Example

r = ReverseList([1, 2, 3])
print(r.reverse())
```
