```python
class Solution():
    def romanToInt(self,s :str):
        roman = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        total = 0

        for i in range(len(s)):
            if i < len(s) - 1 and roman[s[i]] < roman[s[ i + 1]]:
                total -= roman[s[i]]
            else:
                total += roman[s[i]]
        return total


sol = Solution()
print(sol.romanToInt("III"))      # 3
print(sol.romanToInt("LVIII"))    # 58
print(sol.romanToInt("MCMXCIV"))  # 1994

```
